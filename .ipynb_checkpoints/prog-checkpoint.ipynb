```python
import pandas as pd
from io import StringIO
import numpy as np
import string
```

```python
TESTDATA = StringIO(data)
```

```python
parser=pd.read_csv(TESTDATA,sep='\t',names=['section'])
```

```python
# LeadingAsterisk 
# leadingNumeral 
# endsInPeriod
# leadingSpace
# stringLength
# percentCaps
# ellipses
```

```python
def startWithRoman (Instr):
    """
    this function return true if the given string starts with Simple Roman numeral followed by period or ' '
    """
    return re.match('^(?=[MDCLXVI])M*(C[MD]|D?C*)(X[CL]|L?X*)(I[XV]|V?I*)[ |.]',Instr) is not None

#######################################
def leadingWhiteSpace(Instr):
    """
    this function return true if the given string starts with WhiteSpace character
    """
    return re.match('^\s',Instr) is not None

########################################
def ellipses(Instr):
    """
    this function return true if the given string contains ellipses
    """
    return re.search(r'(\w+)\.{3,}',Instr) is not None 
```

```python
#removing nan values
parser = parser.dropna()
```

```python
# LeadingAsterisk
parser['LeadingAsterisk'] = parser['section'].apply(str.startswith, args='*') 

    # # leading arabic numeral
ArabicNumeral= parser['section'].str.lstrip(' ').str[0].apply(str.isdigit)
    # # leading Roman numeral
RomanNumerals= parser['section'].str.lstrip(' ').apply(lambda st: startWithRoman(st))
# leadingNumeral 
parser['LeadingNumeral'] = ArabicNumeral | RomanNumerals

# endsInPeriod
parser['endsInPeriod'] = parser['section'].apply(str.endswith, args='.') 

# leadingWhiteSpace
parser['leadingWhiteSpace'] = parser['section'].apply(lambda st: leadingWhiteSpace(st))
  
```

```python
#  # now we can remove white spaces in the section
    #remove spaces in begening and end of sections
parser['section'] = parser['section'].map(lambda x: x.lstrip(' ').rstrip(' ')) 
```

```python
# stringLength
parser['stringLength'] = parser['section'].apply(len) 
```

```python
#  #remove page's numbers
parser = parser.drop(parser[parser['stringLength']<4].index)
```

```python
# percentCaps #presentation precision = 2
parser['percentCaps'] = parser['section'].apply(lambda st: round(sum(1 for c in st if c.isupper())*100/len(st),2))
# ellipses
parser['ellipses'] = parser['section'].apply(lambda st: ellipses(st))
```
